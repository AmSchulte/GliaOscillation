```python
plt.figure(figsize=(16,8))
plt.plot(analyse.avg_signal_er)
plt.scatter(analyse.experiment_start_er, analyse.avg_signal_er[analyse.experiment_start_er], c="r")

plt.plot(analyse.avg_signal_cy)
plt.scatter(analyse.experiment_start_cy, analyse.avg_signal_cy[analyse.experiment_start_cy], c="g")

(analyse.experiment_start_cy - analyse.experiment_start_er)*0.859
```

```python
lag_peaks = []
for cell in analyse.cells:

    print(cell.oscillation)

    
    #peaks_cy, prominences = find_peaks(cell.cy.signal, prominence=0.5)
    
    plt.figure(figsize=(16,8))
    plt.plot(cell.time, cell.er.signal, c='xkcd:dark blue', alpha=0.5)
    plt.plot(cell.time, cell.er.lowpass, c='xkcd:dark blue')

    plt.plot(cell.time, cell.cy.signal, c='xkcd:slate blue', alpha=0.5)
    plt.plot(cell.time, cell.cy.lowpass, c='xkcd:slate blue')

    
    #plt.scatter(cell.time[analyse.experiment_start_er], cell.er.signal[analyse.experiment_start_er], c="r")
    #plt.scatter(cell.time[analyse.experiment_start_cy], cell.cy.signal[analyse.experiment_start_cy], c="g")
    
    #plt.scatter(cell.time[peaks_cy], cell.cy.signal[peaks_cy], c="g")

    try:
        cell.calculate_lag_flavors()
        high_var = np.where(cell.variance_er>0.0015)[0]
        true_matches = []
        true_lags = []
        for match, lag in zip(cell.cy_peak.matches, cell.cy_peak.lag):        
            if match[1] in high_var:
                true_matches.append(match)
                true_lags.append(lag)
                
        lags = []
        if len(true_matches) > 1:
            for match,lag in zip(true_matches,true_lags):
                x = [cell.time[match[1]], cell.time[match[0]]]
                y = [cell.er.lowpass[match[1]], cell.cy.lowpass[match[0]]]
                plt.plot(x, y, 'go-', c='C2')
                lags.append(lag) 
            mean_lag = np.mean(lags)
            lag_peaks.append(mean_lag)
    except:
        pass

    print(cell.mean_lag)

    plt.show()
print(analyse.mean_lags)    
```

```python
lag_peaks_all = []
for i in range(len(messungen)):
    #read data
    messung = i
    er = ana.ReadData(directory=messungen[messung]["er"])
    cy = ana.ReadData(directory=messungen[messung]["cy"])
    analyse = ana.CellData(er=er, cy=cy)
    
    #get frequencies of each cell
    lag_peaks = analyse.mean_lags
    lag_peaks = [l for l in lag_peaks if l is not None]
    lag_peaks_all.extend(lag_peaks)   
```

```python
labels=['Lag peaks']
name = 'Lag peaks'

c='k'
fig, ax = plt.subplots(figsize=(5,8))
alpha=1
standard_props = {"color":c, "alpha":alpha, "linewidth":5}
markers= ['D','o']
line_props = standard_props
bbox_props = {**standard_props, 'facecolor': 'w', 'edgecolor':c} #dict(edgecolor=c, alpha=alpha, linewidth=5, facecolor='w')
median_props = standard_props
cap_props = standard_props
linewidth = 5
fontsize = 32

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines['left'].set_linewidth(linewidth)
ax.spines['bottom'].set_linewidth(linewidth)
ax.tick_params(length=10, width=linewidth)
#ax.set_xlim([0, 100])
ax.set_ylim([0, 18])
plt.xticks(fontname="Arial", fontsize=fontsize)
plt.yticks(fontname="Arial", fontsize=fontsize)


plt.ylabel('Lag (s)', fontname="Arial", fontsize=34)


sns.boxplot(data=lag_peaks_all, width=0.5, palette='colorblind', whiskerprops=line_props, boxprops=bbox_props
            ,capprops=cap_props, medianprops=median_props, fliersize=0)
sns.stripplot(data=lag_peaks_all, ax=ax, dodge=True, jitter=0.2, size=10, marker='o', color='grey')
ax.set_xticklabels(labels, fontname="Arial", fontsize=34)


#plt.savefig(name+'.png', bbox_inches = "tight")

plt.show()
```
