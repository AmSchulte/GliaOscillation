```python
messung = 0
er = ana.ReadData(directory=messungen[messung]["er"])
cy = ana.ReadData(directory=messungen[messung]["cy"])
analyse = ana.CellData(er=er, cy=cy, fps=0.375, frames=messungen[messung]["frames"])
```

```python
plt.figure(figsize=(16,8))
plt.plot(analyse.avg_signal_er)
plt.scatter(analyse.experiment_start_er, analyse.avg_signal_er[analyse.experiment_start_er], c="r")

plt.plot(analyse.avg_signal_cy)
plt.scatter(analyse.experiment_start_cy, analyse.avg_signal_cy[analyse.experiment_start_cy], c="g")

(analyse.experiment_start_er - analyse.experiment_start_cy)*0.375
```

```python
cells_no_oscillation_before = []
for cell in analyse.cells:

    #np.where((cell.cy.derivative>0)&(cell.cy.derivative>0))
    print(cell.oscillation)
    
    if 'before' not in cell.oscillation:
        cells_no_oscillation_before.append(cell)
    
    peaks_cy, prominences = find_peaks(cell.cy.signal, prominence=0.5)
    #print(peaks_cy)
    
    plt.figure(figsize=(16,8))
    plt.plot(cell.er.signal)
    plt.plot(cell.cy.signal)
    
    #plt.plot(cell.er.lowpass[200:600])
    #plt.plot(cell.cy.lowpass[200:600])
    
    
    #plt.scatter(peaks_cy, cell.cy.signal[peaks_cy], c="g")

    
    if 'before' not in cell.oscillation:
    # calcualte fidulcial point
        point_a = 0
        point_b = np.argmin(cell.er.signal)
        print(point_b)
        reaction_start_er = np.argmax(detrend(cell.er.signal[point_a:point_b]))
        
        point_a = 0
        point_b = np.argmax(cell.cy.signal)
        reaction_start_cy = np.argmin(detrend(cell.cy.signal[point_a:point_b]))

        difference = (reaction_start_er - reaction_start_cy)*0.375
        print('Lag: '+ str(difference))
        plt.scatter(reaction_start_er, cell.er.signal[reaction_start_er], c="r")
        plt.scatter(reaction_start_cy, cell.cy.signal[reaction_start_cy], c="g")
        
    plt.show()
```

```python
new_mean_er = np.mean([cell.er.signal for cell in cells_no_oscillation_before], axis=0)
new_mean_cy = np.mean([cell.cy.signal for cell in cells_no_oscillation_before], axis=0)

point_a = 100
point_b = np.argmin(new_mean_er)
#point_b = 158
experiment_start_er = np.argmax(detrend(new_mean_er[point_a:point_b]))+100
print(point_b)
plt.plot(detrend(new_mean_er[point_a:point_b]))

point_a = 0
point_b = np.argmax(new_mean_cy)
print(point_b)
experiment_start_cy = np.argmin(detrend(new_mean_cy[point_a:point_b]))
```

```python
plt.figure(figsize=(16,8))
plt.plot(new_mean_er)
plt.scatter(experiment_start_er, new_mean_er[experiment_start_er], c="r")

plt.plot(new_mean_cy)
plt.scatter(experiment_start_cy, new_mean_cy[experiment_start_cy], c="g")

(experiment_start_er - experiment_start_cy)*0.375
```

```python
lags = []
for i in range(len(messungen)):
    messung = i
    er = ana.ReadData(directory=messungen[messung]["er"])
    cy = ana.ReadData(directory=messungen[messung]["cy"])
    analyse = ana.CellData(er=er, cy=cy, fps=0.375, frames=messungen[messung]["frames"])
    
    plt.figure(figsize=(16,8))
    plt.plot(analyse.avg_signal_er)
    plt.scatter(analyse.experiment_start_er, analyse.avg_signal_er[analyse.experiment_start_er], c="r")

    plt.plot(analyse.avg_signal_cy)
    plt.scatter(analyse.experiment_start_cy, analyse.avg_signal_cy[analyse.experiment_start_cy], c="g")
    plt.show()
    
    plt.figure(figsize=(16,8))
    plt.xlim([analyse.experiment_start_er-50,analyse.experiment_start_er+50])
    plt.plot(analyse.avg_signal_er, marker='x')
    plt.scatter(analyse.experiment_start_er, analyse.avg_signal_er[analyse.experiment_start_er], c="r")

    plt.plot(analyse.avg_signal_cy, marker='x')
    plt.scatter(analyse.experiment_start_cy, analyse.avg_signal_cy[analyse.experiment_start_cy], c="g")
    plt.show()
    
    lag = (analyse.experiment_start_er - analyse.experiment_start_cy)*0.375
    print(lag)
    lags.append(lag)
```
